```python
import numpy as np

array_2x2 = np.array([[5, 2],
                      [7, 4]])
print(array_2x2)
```

```python
random_array = np.random.rand(3,4)
print(random_array)
```

```python
ones_array = np.ones((3, 3))
zeros_array = np.zeros((3, 3))

print("Array of ones:\n", ones_array)
print("\nArray of zeroes:\n", zeros_array)
```

```python
random_array2 = np.random.rand(6, 3)
print("Original array:\n", random_array2)

transposed = random_array2.T
print("\nTransposed array:\n", transposed)
```

```python
n, p, m = 2, 3, 4
A = np.random.rand(n, p)
B = np.random.rand(p, m)

dot_product = np.dot(A, B)

print("Matrix A:\n", A)
print("\nMatrix B:\n", B)
print("\nDot product:\n", dot_product)
```
